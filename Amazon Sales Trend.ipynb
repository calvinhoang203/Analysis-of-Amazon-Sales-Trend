```python
# Question 2: What are the top-rated products based on user ratings?
```

```python
convert_rating_count = [] # an empty list
for index, row in amazon_data.iterrows():
    convert_rating_count.append(str(row['rating_count']).replace(",",""))  # remove the comma
convert_rating_count = [float(i) for i in convert_rating_count]
amazon_data['convert_rating_count'] = convert_rating_count
```

```python
# Bar plot for the top-rated products based on user ratings
top_5.plot(kind='bar', color='red')
plt.title('Top-rated Products Based on User Ratings')
plt.xlabel('Product Names')
plt.ylabel('Ratings')
plt.show()
```

The cable has more rating counts compared to the in-ear headphones, with a count of 853,945.
```python
# Question 3: Is there a correlation between the number of ratings and the average rating of a product?
```

- There are ratings that only have '|' string, meaning that we need to remove rows containing only those strings.

```python
convert_rating = [] # an empty list
for index, row in amazon_data.iterrows():
    if row['rating'] == '|':
        row['rating'] = 0
        convert_rating.append(str(row['rating']))
    else:
        convert_rating.append(str(row['rating']))
convert_rating = [float(i) for i in convert_rating]
amazon_data['convert_rating'] = convert_rating
```

```python
# Calculate the average rating and total rating count for each product
product_ratings = amazon_data.groupby('product_id').agg({'convert_rating': 'mean', 'convert_rating_count': 'sum'}).reset_index()

# Plotting the correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(x='convert_rating', y='convert_rating_count', data=product_ratings)
plt.title('Correlation between Number of Ratings and Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Number of Ratings')
plt.show()
```

- Based on the graph, there seems to be a positive correlation between the number of ratings and average rating of a product. A positive correlation would suggest that products with a higher number of ratings tend to have higher average ratings, indicating higher overall satisfaction. Thus, I strongly believe that Amazon provide very reliable products to their customers.

```python
# Question 4: What are top products that have the most discounted price?
```

- Since the currency in the dataset is in Rupee, I will convert it into Dollar currency with the current market price. (1 ₹ = 0.012 USD)

```python
# Convert the type of 'actual_price' column from string to float
convert_actual_price_USD = [] # an empty list
for index, row in amazon_data.iterrows():
    convert_actual_price_USD.append(str(row['actual_price']).replace(',', '').replace('₹', '')) # Remove both '₹' and ','
convert_actual_price_USD = [float(i) for i in convert_actual_price_USD]
convert_actual_price_USD = [price * 0.012 for price in convert_actual_price_USD] # convert prices to dollars using the exchange rate (0.012)
amazon_data['convert_actual_price_USD'] = convert_actual_price_USD
amazon_data.iloc[0:10, [0,1,18]]
```

```python
# Convert the type of 'discounted_price' column from string to float
convert_discounted_price_USD = [] # an empty list
for index, row in amazon_data.iterrows():
    convert_discounted_price_USD.append(str(row['discounted_price']).replace(',', '').replace('₹', '')) # Remove both '₹' and ','
convert_discounted_price_USD = [float(i) for i in convert_discounted_price_USD]
convert_discounted_price_USD = [price * 0.012 for price in convert_discounted_price_USD] # convert prices to dollars using the exchange rate (0.012)
amazon_data['convert_discounted_price_USD'] = convert_discounted_price_USD
amazon_data.iloc[0:10, [0,1,19]]
```

```python
# Calculate the discount amount for each product
amazon_data['discount_amount'] = amazon_data['convert_actual_price_USD'] - amazon_data['convert_discounted_price_USD']

# Find the top products with the most discounted prices
top_discounted_products = amazon_data.sort_values(by='discount_amount', ascending=False).head(10)
top_discounted_products[['product_name', 'category', 'discount_amount']]
```

```python
# Create a bar chart displaying the top products with the most discounted prices
plt.figure(figsize=(12, 6))
plt.bar(top_discounted_products['product_name'], top_discounted_products['discount_amount'], color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Discount Amount')
plt.title('Top Products with the Most Discounted Prices in USD')
plt.xticks(rotation=90)  # Adjust rotation for better readability
plt.show()
```

- Based on the data, we can see that the Sony Bravia 164 cm (65 inches) 4K Ultra HD Smart LED TV is the most discounted product by 742.920 dollars.  

```python
# Question 5: What is the correlation between the discount percentage and the number of ratings?
```

```python
# Convert the type of 'discount_percentage' column from string to float
convert_discount_percentage = [] # an empty list
for index, row in amazon_data.iterrows():
    convert_discount_percentage.append(str(row['discount_percentage']).replace('%', '')) # Remove both '₹' and ','
convert_discount_percentage = [float(i) for i in convert_discount_percentage]
amazon_data['convert_discount_percentage'] = convert_discount_percentage
amazon_data.head(5)
```

```python
# Calculate the discount percentage and total rating count for each product
discounted_product_ratings = amazon_data.groupby('product_id').agg({'convert_discount_percentage': 'mean', 'convert_rating_count': 'sum'}).reset_index()

# Plotting the correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(x='convert_discount_percentage', y='convert_rating_count', data=discounted_product_ratings)
plt.title('Correlation between Discount Percentage and Number of Ratings')
plt.xlabel('Discount Percentage')
plt.ylabel('Number of Ratings')
plt.show()
```

```python
# Question 10: Are there certain products that consistently receive 'Good' or 'Bad' reviews?
```

- Let's assume that the products that consistently receive 'Good' reviews are the ones that have more than 4 star rating and the highest number of ratings, while the products that consistently receive 'Bad' reviews are the ones that have less than 4 star rating and the highest number of ratings.

```python
good_reviews = amazon_data[amazon_data['convert_rating'] >= 4]
good_reviews = good_reviews.sort_values(by = 'convert_rating_count', ascending=False)
certain_products_with_good_reviews = good_reviews.iloc[0:10,[0,1,18,16]]
certain_products_with_good_reviews
```

```python
# Bar plot for products with good reviews
sns.barplot(x='convert_rating_count', y='product_name', data=certain_products_with_good_reviews, palette='viridis')
plt.title('Top Products with Good Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Product Name')
plt.show()
```

```python
bad_reviews = amazon_data[amazon_data['convert_rating'] < 4]
bad_reviews = bad_reviews.sort_values(by = 'convert_rating', ascending=True)
bad_reviews = bad_reviews.sort_values(by = 'convert_rating_count', ascending=False)
certain_products_with_bad_reviews = bad_reviews.iloc[0:10,[0,1,18,16]]
certain_products_with_bad_reviews
```

- 

```python
# Bar plot for products with bad reviews
sns.barplot(x='convert_rating_count', y='product_name', data=certain_products_with_bad_reviews, palette='viridis')
plt.title('Top Products with Bad Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Product Name')
plt.show()
```
