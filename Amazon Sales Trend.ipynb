```python
convert_rating_count = []
for index, row in amazon_data.iterrows():
    convert_rating_count.append(float(row['rating_count'].replace(",","")))  # remove the comma
amazon_data['convert_rating_count'] = convert_rating_count
amazon_data.groupby('product_name').sum().sort_values(by='convert_rating_count', ascending = False) # Group the number of products in descending order
```
