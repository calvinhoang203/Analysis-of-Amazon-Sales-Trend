```python
# Since there are some missing values in the rating_count column, it can cause errors when we use that column for analysis.
# Thus, we need to remove them.
```

```python
# Question 2: What are the top-rated products based on user ratings?
```

```python
convert_rating_count = [] # an empty list
for index, row in amazon_data.iterrows():
    convert_rating_count.append(str(row['rating_count']).replace(",",""))  # remove the comma
convert_rating_count = [int(i) for i in convert_rating_count]
amazon_data['convert_rating_count'] = convert_rating_count
top_rated_products = amazon_data.groupby('product_name').sum().sort_values(by='convert_rating_count', ascending = False) # Group the number of products in descending order
top_5 = top_rated_products.iloc[0:5,[0,1,15]]
top_5
```

```python
# Bar plot
top_5.plot(kind='bar', color='red')
plt.title('Top-rated Products Based on User Ratings')
plt.xlabel('Product Names')
plt.ylabel('Ratings')
plt.show()
```

```python
# Question 3: Which category has the highest number of products?
```

```python
category_counts = amazon_data['category'].value_counts() # Count the number of products in each category
top_5_categories = category_counts.head(5)
top_5_categories
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='category', data=amazon_data, order=top_5_categories.index)
plt.xticks(rotation = 'vertical')
plt.title('Number of Products in The Top 5 Categories')
plt.plot(top_5_categories.sort_values(ascending=False), color = 'red')
plt.show()
```

```python
# Question 4: What are the top 5 most popular products in each category based on the number of reviews?
```

```python
usb_cables = amazon_data[amazon_data['category'] == 'Computers&Accessories|Accessories&Peripherals|Cables&Accessories|Cables|USBCables']
usb_cables = usb_cables.drop(columns=['convert_rating_count'])
```

```python
convert_rating_count_usb_cables = [] # an empty list
for index, row in usb_cables.iterrows():
    convert_rating_count_usb_cables.append(str(row['rating_count']).replace(",",""))  # remove the comma
convert_rating_count_usb_cables = [int(i) for i in convert_rating_count_usb_cables]
usb_cables['convert_rating_count_usb_cables'] = convert_rating_count_usb_cables
top_5_usb_cables = usb_cables.groupby('product_name').sum().sort_values(by='convert_rating_count_usb_cables', ascending = False)
top_5 = top_5_usb_cables.iloc[0:5,[0,1,15]]
top_5
```

```python
# Pie chart for top 5 products in USB Cables category
usb_cables_pie_chart = usb_cables.groupby('product_name')['convert_rating_count_usb_cables'].sum()
top_5_usb_cables = usb_cables_pie_chart.sort_values(ascending=False).head(5)
plt.title('Top 5 USB Cables Sold at Amazon')
plt.pie(top_5_usb_cables, labels = top_5_usb_cables.index, autopct='%1.1f%%')
plt.show()
```
